```python
text = "I'm working as a Data Scientist in the U.S. and Earning $14000."
```

```python
print(text.split())
```

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(text)
```

```python
doc
```

```python
type(doc)
```

```python
print([token for token in doc])
```

```python
[type(token) for token in doc]
```

```python
print([token.text for token in doc])
```

```python
# Comparing whitespaces and SpaCy tokenizer
print(text.split())
print([token.text for token in doc])
```

```python
print(len(text.split()))
print(len([token.text for token in doc]))
```
