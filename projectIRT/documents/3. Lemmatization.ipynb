```python
text = 'The sky is clear and the stars are twinkling.'
```

```python
import spacy 
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(text)
```

```python
[(token.text, token.lemma_) for token in doc]
```
